```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a, b = 5., 3.
n = 65
mr = np.empty((n,n))
for x in range(n):
    for y in range(n):
        cx = -4 + (x / n) * 8
        cy = -4 + (y / n) * 8
        d = max(abs(cx) - a/2, abs(cy) - b/2)
        mr[y,x] = d
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(mr, cmap='RdBu')
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(abs(mr)<0.07, cmap='Greys')
```

```python
import math
radius = 2.2
mc = np.empty((n,n))
for x in range(n):
    for y in range(n):
        cx = -4 + (x / n) * 8
        cy = -4 + (y / n) * 8
        d = math.sqrt(cx**2 + cy**2) - radius
        mc[y,x] = d
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(mc, cmap='RdBu')
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(abs(mc)<0.07, cmap='Greys')
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(np.minimum(mr, mc), cmap='RdBu')
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(np.maximum(mr, mc), cmap='RdBu')
```

```python
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(abs(np.maximum(mr, mc))<0.07, cmap='Greys')
```
