A point is a location in 3D space, represented by an `X`, `Y`, and `Z` coordinate.

## Point object

COMPAS provides a basic point object that supports basic mathemetical operations and provides easy access to basic algorithms.

https://compas-dev.github.io/main/api/generated/compas.geometry.Point.html

```python
from compas.geometry import Point
```

```python
a = Point(0, 0, 0)
b = Point(1, 0, 0)
c = Point(1, 1, 0)
d = Point(0, 1, 0)
```

## Basic math

Point objects in COMPAS support basic mathematical operations, such as addition, subtraction, multiplication with a scalar (scaling).

```python
b - a
```

```python
a + b
```

```python
c * 0.5
```

> Note that the opposite operation (`0.5 * c`) is not supported!

## Plotting

```python
from compas_plotters import Plotter
```

Points can be visualized using a basic plotter and its `draw_points` function. This function expects as input a list of dictionaries providing predefined visualization information. Many of the `keys` in the dictionaries are optional, only `pos` and `radius` are mandatory.

> Points are drawn as circles. Therefore, the `radius` of the point is the actual `radius` of that circle. Since COMPAS plots are automatically scaled to fit their contents, drawing a single point will fill the entire plot...

```python
plotter = Plotter()
plotter.draw_points([{'pos': [0, 0, 0], 'radius': 0.1}])
plotter.show()
```

### Styling

The point dictionaries support several additional styling `keys`:
* facecolor: the color of the face of the circle (hex or rgb). default is `'#ffffff'` (white).
* edgecolor: the color of the edge of the circle (hex or rgb). default is `'#000000'` (black).
* edgewidth: the line width of the edge of the circle. default is `1.0`.

### Labeling

There are also a few text options:
* text: the text string of the label.
* fontsize: the size of the text. default is `12`.
* textcolor: the color of the text. default is black.

```python
plotter = Plotter()
plotter.draw_points([
    {'pos': a, 'radius': 0.05, 'text': 'a', 'fontsize': 16},
    {'pos': b, 'radius': 0.05, 'text': 'b', 'fontsize': 24},
    {'pos': c, 'radius': 0.05, 'text': 'c', 'fontsize': 8},
    {'pos': d, 'radius': 0.05, 'text': 'd'}
])
plotter.show()
```
